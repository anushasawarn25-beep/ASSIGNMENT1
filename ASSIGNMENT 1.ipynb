5 What is mutability in Python ?
In Python, mutability refers to whether an object’s state or content can be changed after it is created. Python objects are broadly classified into two categories based on this characteristic:
Mutable objects: Their contents can be modified after creation.
Immutable objects: Once created, their content cannot be altered.
Mutable Objects
These objects allow you to change their contents without creating a new object. Common mutable types include:
Lists (e.g., [1, 2, 3])
Dictionaries (e.g., {'key': 'value'})
Sets (e.g., {1, 2, 3})

Immutable Objects
These objects cannot be changed once created. Attempting to change them results in the creation of a new object rather than modifying the existing one. Common immutable types include:
Integers (e.g., 5)
Floats (e.g., 3.14)
Strings (e.g., "hello")
Tuples (e.g., (1, 2, 3))

 6 Why are lists mutable, but tuples are immutable?
 lists are designed to be flexible — you can add, remove, or change items whenever you want. This makes them ideal for cases where your data needs to evolve over time. On the other hand, tuples are meant to hold data that shouldn’t change once it’s been created. Because of this, tuples are immutable, meaning you can’t alter their contents after making them.
Here’s why Python makes this distinction:
Different Purposes
Lists serve as containers you can easily update, grow, or shrink. Tuples are like fixed records — perfect when you want to guarantee the data stays the same throughout your program, like coordinates or configuration settings.
How They’re Built Under the Hood
Lists are implemented as dynamic arrays with extra space set aside to handle changes efficiently. Tuples, by contrast, are fixed in size and simpler. This makes tuples faster and more memory-friendly.
Hashing and Usage as Dictionary Keys
Because tuples never change, they can be safely used as keys in dictionaries or stored in sets. Lists can’t because their content might change, breaking internal data structures that depend on consistent keys.

11.What is the purpose of conditional statements in Python ?
Conditional statements  are fundamental building blocks used to control the flow of a program by making decisions based on certain conditions. Their primary purpose is to allow the program to execute different blocks of code depending on whether a condition evaluates to True or False.
What Are Conditional Statements?
In Python, the common conditional statements are if, elif (short for else if), and else. These statements enable branching logic:
if: Checks a condition. If the condition is true, executes the indented block of code beneath it.
elif: (optional) Checks another condition if the previous if or elif was false.
else: (optional) Executes a block of code if none of the previous conditions were true.
Why Are Conditional Statements Important?
Decision Making: They help the program decide which path to take based on inputs, variables, or logical conditions.
Control Flow: Conditional statements make programs dynamic and capable of handling varying scenarios instead of executing all statements sequentially.
Improve Code Flexibility: By testing different conditions, your code can behave differently, making it more versatile.
EXAMPLE:
age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")

12 How does the elif statement work ?
When Python encounters an if-elif-else block, it evaluates the conditions one by one from top to bottom:
if condition is evaluated first.
If it's True, the block of code under the if runs, and the whole conditional ends there—no further conditions are checked.
If the if condition is False, Python moves to the first elif condition (if any).
If this elif condition is True, its block executes, and the rest are skipped.
If elif is also False, Python continues checking any subsequent elifs in order.
If no if or elif condition is True, the optional final else block will be executed.
EXAMPLE:
temperature = 25

if temperature > 30:
    print("It's a hot day.")
elif temperature > 20:
    print("The day is warm.")
elif temperature > 10:
    print("It's a bit chilly.")
else:
    print("It's cold.")

13 What is the difference between for and while loops ?
A for loop is typically used when you know in advance how many times you want to repeat a block of code. It iterates over a sequence, such as a range of numbers. A while loop continues executing as long as a certain condition remains true, and is useful when the number of iterations is not fixed beforehand.
EXAMPLE:
   FOR LOOP
for i in range(1, 6):
    print(i)
    i = 1
  WHILE LOOP
while i <= 5:
    print(i)
    i += 1

For loops are generally preferred when the number of iterations is known before entering the loop (e.g., iterating over a fixed range or collection).
While loops are better when the number of iterations is uncertain and the loop depends on a dynamic condition evaluated during execution.

Since you don't know in advance how many attempts the user will take, you cannot use a for loop effectively (unless you set an arbitrary max).
```python
9#. Write a Python program to convert user input from string to integer, float, and boolean types

user_input = input("Enter a value: ")

try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")

try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")


if user_input.strip().lower() in ["false", "0", "no", "none"]:
    bool_value = False
else:
    bool_value = True

print("Boolean value:", bool_value)
```

```python
#10. Write code to demonstrate type casting with list elements

str_numbers = ['10', '20', '30', '40']

int_numbers = [int(num) for num in str_numbers]

float_numbers = [float(num) for num in int_numbers]

back_to_strings = [str(num) for num in float_numbers]

print("Original strings:", str_numbers)
print("Converted to integers:", int_numbers)
print("Converted to floats:", float_numbers)
print("Converted back to strings:", back_to_strings)
```

```python
 #15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number to calculate its factorial: "))

factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1

print(f"The factorial of {num} is {factorial}")
```
